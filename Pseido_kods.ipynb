```python
# Algoritma apraksts (pseido-kods)

Šajā sadaļā aprakstīts programmas darbības algoritms vienkāršotā pseido-koda veidā.

Programma sākas ar datu ielādi no JSON faila. Ja fails neeksistē, tiek izveidots tukšs izdevumu saraksts.

Pēc tam programma nepārtraukti rāda izvēlni, līdz lietotājs izvēlas iziet.

Pseido-kods:

START

ielādē izdevumus no JSON faila

WHILE programma darbojas

    parādi izvēlni

    nolasīt lietotāja izvēli

 IF izvēle = 1
    pievienot jaunu izdevumu
    ievadīt datumu
    ievadīt summu
    izvēlēties kategoriju
    ievadīt aprakstu
    pievienot izdevumu sarakstam
    saglabāt JSON failā

ELSE IF izvēle = 2
    parādīt visus izdevumus tabulas formā

ELSE IF izvēle = 3
    ievadīt mēnesi
    atlasīt izdevumus ar šo mēnesi
    parādīt filtrēto sarakstu

ELSE IF izvēle = 4
    aprēķināt summu katrai kategorijai
     parādīt rezultātu

ELSE IF izvēle = 5
    parādīt izdevumu sarakstu ar numuriem
    lietotājs ievada dzēšamo numuru
     dzēst izvēlēto izdevumu
    saglabāt JSON failā

ELSE IF izvēle = 6
    eksportēt izdevumus CSV failā

ELSE IF izvēle = 7
     beigt programmu

END WHILE

END
```
